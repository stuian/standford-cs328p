# 数据标注

- 有标注数据：半监督学习
- 没有标注数据但有钱：众包标注
- 既没有标注数据又没有钱：弱监督

## 半监督学习

只有少量标注数据，比如生活中，只有很少部分的用户进行了反馈。

数据分布的假设

- 相似的样本具有相同的标注
- 同一簇内的样本具有相同的标注
- 流型假设：数据在低维空间分布均匀

### 自训练

## 众包标注

找很多人（不一定是专家）来标注，低成本的。

- 设计相当于简单的任务，如何标注
- 成本控制
- 质量控制

### 成本控制

主动学习

> 减少标注的数量

- 不确定性采样：把不确定的样本给人来标注
- 多模型投票选出难的样本

自训练+主动学习

### 质量控制

- 让多个人进行标注，最后投票
- 或先发给两个人，如果两个人意见相同，就不用发给其他人
- 发给一个人，和模型预测进行对比
- 对标注工的准确率进行评估

## 弱监督 

- 写一堆规则，去产生一堆弱标签，最终汇总投票
- 例：图片标注。把图片让各大公司的图片分类工具去分类。
